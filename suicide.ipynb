```python
# Dependencies and Setup
import pandas as pd
import numpy as np

od_to_load = "od.csv"
suicide2k17_to_load ="SUICIDE2017.csv"
suicide2k16_to_load ="SUICIDE2016.csv"
suicide2k15_to_load ="SUICIDE2015.csv"

od_data = pd.read_csv(od_to_load)
suicide2k17 = pd.read_csv(suicide2k17_to_load)
suicide2k16 = pd.read_csv(suicide2k16_to_load)
suicide2k15 = pd.read_csv(suicide2k15_to_load)


suicide2k17
```

```python
suicide2k17 = suicide2k17[["STATE","RATE","DEATHS"]]

suicide2k17 = suicide2k17.rename(columns={"STATE": "state_name",
                                          "RATE": "suicide_rate_2017",
                                          "DEATHS": "suicide_deaths_2017"})

# suicide2k17 = suicide2k17.set_index("state_name")

suicide2k17
```

```python
suicide2k16 = suicide2k16[["STATE","RATE","DEATHS"]]

suicide2k16 = suicide2k16.rename(columns={"STATE": "state_name",
                                          "RATE": "suicide_rate_2016",
                                          "DEATHS": "suicide_deaths_2016"})

# suicide2k16 = suicide2k16.set_index("state_name")

suicide2k16
```

```python
suicide2k15 = suicide2k15[["STATE","RATE","DEATHS"]]

suicide2k15 = suicide2k15.rename(columns={"STATE": "state_name",
                                          "RATE": "suicide_rate_2015",
                                          "DEATHS": "suicide_deaths_2015"})

# suicide2k15 = suicide2k15.set_index("state_name")

suicide2k15
```

```python
suicide_1516 = pd.merge(suicide2k15, suicide2k16, how="left", on=["state_name","state_name"])

suicide_1516

suicide2k15_to_2k17 = pd.merge(suicide_1516, suicide2k17, how="left", on=["state_name","state_name"])

suicide2k15_to_2k17

suicide2k15_to_2k17 = suicide2k15_to_2k17.rename(columns={"state_name":"State"})

suicide_complete = suicide2k15_to_2k17.set_index("State")

suicide_complete
```

```python
suicide_complete.to_csv("suicide_complete.csv")
```
