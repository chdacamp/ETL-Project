```python
import pandas as pd
```

```python
linkList = ['https://www.bls.gov/lau/lastrk15.htm','https://www.bls.gov/lau/lastrk16.htm','https://www.bls.gov/lau/lastrk17.htm']

df_15 = pd.read_html(linkList[0])[0]
df_16 = pd.read_html(linkList[1])[0]
df_17 = pd.read_html(linkList[2])[0]

```

```python
df_15.dropna()
```

```python
df_16.dropna()
```

```python
df_17.dropna()
```

```python
lookup_df = pd.read_csv('./states.csv')
lookup_df.head()
```

```python
unemployment_df = df_15.merge(df_16, on="State",suffixes = ["_15","_16"]).merge(df_17, on="State")
```

```python
unemployment_df = unemployment_df.merge(lookup_df,on="State")
unemployment_df = unemployment_df.rename(columns={"Rank":"Rank_17","State":"State_Name","Abbreviation":"State"})
unemployment_df = unemployment_df.set_index("State")
unemployment_df.head()
```

```python
unemployment_df.to_csv("./unemployment_df.csv")
```
